```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=x*2
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, color="g", linestyle=":", linewidth="8")

plt.xlabel("x-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Adding two lines in the same graph.

```python
x=np.arange(1,11)
y1=x*2
y2=x*3
plt.plot(x,y1, color="g", linestyle=":", linewidth="4")
plt.plot(x,y2, color="r", linestyle="-.", linewidth="5")
plt.grid(True)
plt.show()
```
